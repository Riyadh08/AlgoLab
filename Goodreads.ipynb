```python
import pandas as pd

# Read the CSV file (no need for 'openpyxl')
df = pd.read_csv('scraped_ratings.csv')

# Print the contents
# Store the 'user_id' column in a list
user_id_list = df['User_ID'].tolist()

# Store the 'book_id' column in a list
book_id_list = df['Book_ID'].tolist()

# # Store the 'ratings' column in a list
# ratings_list = df['ratings'].tolist()

# Store the 'user_id' column in a list (uniquely)
unique_user_ids = list(set(user_id_list))

# Store the 'book_id' column in a list (uniquely)
unique_book_ids = list(set(book_id_list))

# Print the unique values to check
print("Unique User IDs:", unique_user_ids)
print("Unique Book IDs:", unique_book_ids)


```

```python
import pandas as pd

# Step 1: Create a dictionary mapping each unique user_id to its index
user_mapping = {user_id: index for index, user_id in enumerate(unique_user_ids)}

# Step 2: Convert the dictionary to a DataFrame
user_mapping_df = pd.DataFrame(list(user_mapping.items()), columns=['User_ID', 'Index'])

# Step 3: Save the DataFrame to a CSV file
user_mapping_df.to_csv('usermapping.csv', index=False)

# Print to verify the DataFrame before saving
print(user_mapping_df)
```

```python
# Step 1: Create a dictionary mapping each unique book_id to its index
book_mapping = {book_id: index for index, book_id in enumerate(unique_book_ids)}

# Step 2: Convert the dictionary to a DataFrame
book_mapping_df = pd.DataFrame(list(book_mapping.items()), columns=['Book_ID', 'Index'])

# Step 3: Save the DataFrame to a CSV file
book_mapping_df.to_csv('bookmapping.csv', index=False)

# Print to verify the DataFrame before saving
print(book_mapping_df)
```

```python
import pandas as pd
import numpy as np

# Read the CSV file again
df = pd.read_csv('scraped_ratings.csv')

# Assuming the 'scraped_ratings.csv' has the following columns: 'User_ID', 'Book_ID', 'Rating'
# (If column names are different, adjust accordingly)

# Step 1: Create an empty matrix (DataFrame) with users as rows and books as columns, initialized with NaN
ratings_matrix = pd.DataFrame(np.nan, index=unique_user_ids, columns=unique_book_ids)

# Step 2: Populate the matrix using the ratings from 'scraped_ratings.csv'
for _, row in df.iterrows():
    user_id = row['User_ID']
    book_id = row['Book_ID']
    rating = row['Rating']

    # Get the index of user and book using the mappings
    user_index = user_mapping[user_id]
    book_index = book_mapping[book_id]

    # Place the rating in the correct position in the matrix
    ratings_matrix.loc[unique_user_ids[user_index], unique_book_ids[book_index]] = rating

# Step 3: Save the matrix as a CSV file
ratings_matrix.to_csv('user_book_ratings_matrix.csv')

# Print the matrix to verify
print(ratings_matrix)
```
